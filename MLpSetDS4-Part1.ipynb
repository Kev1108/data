**Problem 3.** 請使用上面的資料，建立一個 DataFrame，可顯示 Problem 2 中的圖表 :

```python
# Problem 3

import  pandas as pd

data3A = [["珍珠奶茶", 50, 68, 56, 78, 45, 56, 89, 94],       
         ["茉香綠茶", 35, 112, 134, 145, 98, 108, 123, 145], 
         ["凍檸檬茶", 40, 66, 63, 48, 68, 62, 76, 88],
         ["金桔檸檬", 45, 46, 42, 37, 48, 56, 66, 69],]

data3B = ["飲品", "價格", "Day1", "Day2", "Day3", "Day4", "Day5", "Day6", "Day7"]

#建立一個 DataFrame  :
df1=pd.DataFrame(data3A, columns=data3B)
df1

```
